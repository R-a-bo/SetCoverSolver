```python
def get_all_data():
    all_data = []
    
    for fp in file_paths:
        array = np.loadtxt(fp[0], delimiter=',', dtype=int).flatten()
        all_data.append((array, fp[1]))
        
    final_data = []
    for data, label in all_data:
        expansion = np.negative(np.ones((limit-data.shape[0],), dtype=int))
        final = np.concatenate((data, expansion))
        final_data.append((final, label))
        
    return final_data
```
